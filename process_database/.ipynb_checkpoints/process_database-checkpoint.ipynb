# Import MySQL csv

```python
import pandas as pd
import hashlib
```

```python
original_df = pd.read_csv("out.csv", sep='\t')
```

```python
original_df
```

# Create definition length column

```python
original_df['definition_len'] = original_df.definition_internalRep.apply(lambda x: len(x))
```

```python
original_df
```

# Keep only the longest definition for every entry

```python
longest_def_df = original_df.groupby('entry_description', as_index=False).apply(lambda x: x.sort_values('definition_len', ascending=False).head(1)).reset_index().drop(['level_0', 'level_1', 'definition_id', 'entry_id', 'definition_len'], axis=1)
```

# Remove definitions that end with ...

```python
longest_def_df['definition_ending'] = longest_def_df.definition_internalRep.apply(lambda x: x[-3:])
```

```python
longest_def_explicative_df = longest_def_df[longest_def_df.definition_ending != '...']
```

```python
longest_def_explicative_df = longest_def_explicative_df.drop(['definition_ending'], axis=1)
```

```python
longest_def_explicative_df
```

# Write processed dataframe to csv file

```python
longest_def_explicative_df.to_csv('formatted.csv', index=False, sep='\t')
```
